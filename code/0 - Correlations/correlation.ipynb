```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

listtoselect = []
data = pd.read_csv('dataset.csv')
data2 = data[['Allele','Drug']]
cleaneddata = pd.get_dummies(data2)
corr = cleaneddata.corr()
#cleaneddata.to_csv('dataset_corr.csv')
corr = corr[269:]
for x in corr.columns:
    y = x.split('_')
    if(y[0] == "Allele"):
        listtoselect.append(x)


corr = corr[listtoselect]

fig = plt.figure(figsize = (45,15))
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
#fig.colorbar(cax)
plt.xticks(rotation=90)
ticksx = np.arange(0,len(corr.columns),1)
ticksy = np.arange(0,len(corr.index),1)
ax.set_xticks(ticksx)
ax.set_yticks(ticksy)
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.index)
corr.to_csv('dataset_corr.csv')
plt.savefig("figure3.pdf")


```

```python
import numpy as np

# Create correlation matrix
corr_matrix = corr.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features 
corr.drop(to_drop, axis=1, inplace=True)
```
