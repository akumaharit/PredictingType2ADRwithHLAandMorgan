```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../3 - Cleaning & Transforming Data/1 cleaned_dataset_splited_output_NOALLELE_NODRUGDESC.csv', index_col = None)
df.drop(columns=['index','Cohort ethnicity','ADR','Pat1','ExpCtrl2'],inplace = True)
```

```python
df_alleles_onehot = pd.get_dummies(df['Allele'], prefix = 'Allele', drop_first = True)
```

```python
df = pd.concat([df, df_alleles_onehot], axis = 1)
df.drop(columns=['Allele'])
df_drug = pd.read_csv('../../1 - Drugs Descriptors/drug_descriptors_dataset.csv', index_col = None)
```

```python
df = pd.merge(df, df_drug, left_on='Drug',right_on='DRUG_NAME', how='inner')
```

```python
df.drop(columns=['Drug','DRUG_NAME','Allele'],inplace = True)
```

```python
no_adr = df['No_ADR']
df.drop(columns=['No_ADR'], inplace = True)
df = pd.concat([no_adr,df], axis = 1)
```

```python
df.iloc[:,0:9].sum(axis=1).sum() == len(df)
```

```python
df['ADR_SJSI'] = df[['ADR_SJSI','ADR_SJSII']].sum(axis=1)
df.drop(['ADR_SJSII'], axis=1, inplace=True)
df.rename(columns={'ADR_SJSI':'ADR_SJS'}, inplace=True)
df
```

```python
df.drop(columns=['ADR_Agranulocytosis','ADR_FDE','ADR_AIU'], inplace=True)
adr = pd.DataFrame(columns=['ADR'])
df
```

```python
adr['ADR'] = df.iloc[:,1:5].sum(axis=1)
df = pd.concat([adr,df], axis = 1)
df
```

```python
df[['ADR','No_ADR']].sum(axis=1).sum() == len(df)
```

```python
df = df[df[['ADR','No_ADR']].sum(axis=1) == 1]
```

```python
df[['ADR','No_ADR']].sum(axis=1).sum() == len(df)
```

```python
df.drop(columns=['No_ADR','ADR_MPeruption','ADR_MPexanthema','ADR_SJS','ADR_TEN'], inplace=True)
df
```

```python
df.to_csv('cleaned_dataset_YESNOADR_noseq_drugdesc.csv',index=False)
```
