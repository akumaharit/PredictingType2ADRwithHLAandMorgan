# ตัด Drugs และ ADR ที่ไม่ใช้ออก ตามที่รีวิวมา

สิ่งที่จะตัดออก
- EGF -> เป็นโปรตีนหาตัวแทนไม่ได้
- Tetanus -> เป็นโปรตีนหาตัวแทนไม่ได้
- Antiepileptics -> ไม่รู้ว่ายากันชักตัวไหน หาตัวแทนไม่ได้ 
- Antibacterial -> หาตัวแทนไม่ได้ จึงตัดออก เพราะไม่รู้จะใส่โครงสร้างอะไร
- NSAID and 'multi-ingredient cold medication' -> หาตัวแทน NSAIDs ไม่ได้ จึงตัดออก เพราะไม่รู้จะใส่โครงสร้างอะไร NSAIDs ค่อนข้างหลากหลาย มีหลายกลุ่ม
- amoxicillin-clavulanate -> มียาสองตัว หาตัวแทนไม่ได้
- co-trimoxazole เพราะมี 2 structure

ADR ตัด
- DIHS
- DILI
- SOC
- DISR
- BIP
- SOSG
- AIA
- AntiHGMCmyopathy
- AMNM
- ALAT
- Jaundice

```python
#ลบ ADR และยาที่บอกว่าจะตัดออก
ADR_toberemoved = ["DIHS","DILI","SOC","DISR","BIP","SOSG","AIA","AntiHGMCRmyopathy","AMNM","ALAT","Jaundice"]
Drug_toberemoved = ["EGF receptor inhibitors","tetanus antitoxin","antibacterial drugs","antiepileptic drugs","NSAID and 'multi-ingredient cold medication'","amoxicillin-clavulanate",'co-trimoxazole']

#เลือก columns ที่เกี่ยวข้อง
filtered_dataset = raw_dataset[['Drug','Allele','Cohort ethnicity','Pat1','ExpCtrl2','Adverse drug reaction']]
filtered_dataset = filtered_dataset.rename(columns={"Adverse drug reaction":"ADR"})
filtered_dataset = filtered_dataset.drop(filtered_dataset[filtered_dataset.ADR.isin(ADR_toberemoved)].index)
filtered_dataset = filtered_dataset.drop(filtered_dataset[filtered_dataset.Drug.isin(Drug_toberemoved)].index)
filtered_dataset
```

# ตัด AIU, FDE, Agranulocytosis, FDE ออก (ข้อมูลน้อยเกินไป) และรวบ SJSI + SJSII

```python
filtered_dataset = filtered_dataset[filtered_dataset['ADR'] != 'AIU']
filtered_dataset = filtered_dataset[filtered_dataset['ADR'] != 'Agranulocytosis']
filtered_dataset = filtered_dataset[filtered_dataset['ADR'] != 'FDE']
filtered_dataset['ADR'] = filtered_dataset['ADR'].replace('SJSI','SJS')
filtered_dataset['ADR'] = filtered_dataset['ADR'].replace('SJSII','SJS')
```
