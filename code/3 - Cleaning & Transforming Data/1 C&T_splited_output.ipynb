```python
import pandas as pd
import numpy as np

filtered_dataset = pd.read_csv('0 filtered_dataset.csv')
filtered_dataset.rename(columns={'Unnamed: 0':'index'}, inplace=True)
filtered_dataset
```

# ทำ onehot encoding กับ column ADR

```python
# ทำ onehot encoding กับ column ADR

onehotted_dataset = filtered_dataset.copy()
onehotted_dataset_copy = onehotted_dataset.copy()
onehotted_dataset = pd.get_dummies(onehotted_dataset, columns = ['ADR'])
onehotted_dataset = pd.concat([onehotted_dataset, onehotted_dataset_copy['ADR']], axis = 1)
onehotted_dataset = onehotted_dataset.fillna(0)
onehotted_dataset = onehotted_dataset.reset_index()
onehotted_dataset
```

# ทำ onehot เชื้อชาติ

```python
onehotted_dataset2 = onehotted_dataset.copy()
onehotted_dataset2_copy = onehotted_dataset.copy()

onehotted_dataset2 = pd.get_dummies(onehotted_dataset2, columns = ['Cohort ethnicity'])
onehotted_dataset2 = pd.concat([onehotted_dataset2, onehotted_dataset2_copy['Cohort ethnicity']], axis = 1)
onehotted_dataset2 = onehotted_dataset2.fillna(0)
onehotted_dataset2
```

# แปลงให้ 1 observation represent 1 คน (โดยอิงจำนวน Observation จาก Pat1)

```python
# เตรียม dataset สุดท้าย
final_dataset = pd.DataFrame(columns = onehotted_dataset2.columns)
final_dataset['No_ADR'] = ""
final_dataset
```

ความหมาย
Pat1 0/20 = 0 ใน 20 คนที่ใช้ยา เกิด ADR แล้วพบยีนที่สนใจ
#ก็คือในคนที่ใช้ยาและมียีน 20 คน มี 0 คนเกิด ADR
Exp2 15/43 = 15 ใน 43 คนที่ใช้ยา แล้วไม่เกิด ADR แต่มียีนที่สนใจ
#ก็คือในคนที่ใช้ยาและมียีน 43 คน มี 15 คนไม่เกิด ADR
Pop3 4/8 = 4 ใน 8 คนพบยีนที่สนใจ

ดังนั้นจำนวน observation จะเท่ากับเลขหน้าของ Pat1 + Exp2

```python
# จัด dataset ใหม่

totpat1_eachrow = {}
totexpctrl2_eachrow = {}

for lab,row in onehotted_dataset2.iterrows():
    numbertorepeat = int(row['Pat1']) #อันนี้คือจำนวนครั้งที่อยากจะ repeat
    totpat1_eachrow = {lab: numbertorepeat}
    for x in range(numbertorepeat):
        toadd = pd.DataFrame(data = [row.values], columns=row.index)
        ADR = "ADR_" + row['ADR']
        toadd = toadd.assign(**{ADR : 1})
        toadd = toadd.assign(**{"No_ADR" : 0})
        final_dataset = pd.concat([final_dataset, toadd])

for lab,row in onehotted_dataset2.iterrows():
    numbertorepeat = int(row['ExpCtrl2']) #อันนี้คือจำนวนครั้งที่อยากจะ repeat
    totexpctrl2_eachrow = {lab: numbertorepeat}
    for x in range(numbertorepeat):
        toadd = pd.DataFrame(data = [row.values], columns=row.index)
        ADR = "ADR_" + row['ADR']
        toadd = toadd.assign(**{ADR : 0})
        toadd = toadd.assign(**{"No_ADR" : 1})
        final_dataset = pd.concat([final_dataset, toadd])
```

```python
final_dataset.reset_index(drop=True,inplace=True)
final_dataset.drop(columns=['level_0'],inplace=True)
final_dataset
```

```python
#check ว่าแปลงแล้วผลรวมถูกต้อง หมายถึง ใน featuers เดียวกันพอทำ onehot เสร็จ รวมกันแล้วจะยังต้องได้จำนวน row เหมือนเดิม!
print(final_dataset[['ADR_AIU','ADR_Agranulocytosis','ADR_FDE','ADR_MPeruption','ADR_MPexanthema','ADR_SJSI','ADR_SJSII','ADR_TEN','No_ADR']].sum().sum())
print(final_dataset[['Cohort ethnicity_Black','Cohort ethnicity_Caucasian','Cohort ethnicity_Diverse','Cohort ethnicity_Mongol']].sum().sum())
```

# Import and Merge Drug's descriptors into dataset by Drugs column

```python
#เตรียม merge descriptors กับยา และ เทียบว่ายาใดถูกตัดออกไปบ้าง และยาใดไม่มีใน smiles

descriptors = pd.read_csv('../1 - Drugs Descriptors/drug_descriptors_dataset.csv')
drugindescriptors = descriptors['DRUG_NAME'].tolist()
drugindescriptors_removedspacebar = [s.replace(" ", "") for s in drugindescriptors]
```

```python
drugincompletedataset = complete_dataset.drop_duplicates('Drug').values
drugincompletedataset = list(drugincompletedataset[:,1])
drugincompletedataset_removedspacebar = [s.replace(" ", "") for s in drugincompletedataset]
```

```python
found = []
notfound = []
for x in drugindescriptors_removedspacebar:
    if x in drugincompletedataset_removedspacebar:
        found.append(x)
    else:
        notfound.append(x)
print(f'Found:{found}')
print(f'Not Found in Complete Dataset (อาจถูกลบออกไปตอนคลีน): {notfound}')
filtered_dataset['Drug'].value_counts()
```

```python
#ลบ spacebar ออกจากทั้ง 2 df ก่อน merge
descriptors['DRUG_NAME'] = descriptors['DRUG_NAME'].str.replace(' ', '')
complete_dataset['Drug'] = complete_dataset['Drug'].str.replace(' ', '')
```
