```python

def value_counts_pandascolumns(dataframe, columnnametocount, cleaned):
    """"รับข้อมูล df, columnsname, cleaned(boolean) แล้ว return เป็น log ไฟล์ว่า หลัง clean แล้วจะมีกี่ observations ของ attribute นั้น"""

    filtered = dataframe[[columnnametocount,'Pat1','ExpCtrl2']].fillna(0)
    count = {}
    for lab,row in filtered.iterrows():
        if row[0] in count:
            count[row[0]] += (int(row[1])+int(row[2]))
        else:
            count[row[0]] = (int(row[1])+int(row[2]))
    if cleaned == True:
        filename = 'log/log_clean_'+ dataframe[columnnametocount].name+ '.txt'
    elif cleaned == False:
        filename = 'log/log_notclean_'+ dataframe[columnnametocount].name + '.txt'

    f = open(filename, 'w')
    #print(pandasseries.value_counts())
    for p,q in count.items():
        f.write(f'{p} is {q}\n')
    f.write(f'witn a total of {sum(count.values())}')
    print(f"Writing to..{filename}..clean status:{cleaned}")
    f.close()
    print("All Done.")
```

```python
value_counts_pandascolumns(raw_dataset, 'Adverse drug reaction', cleaned=False)
value_counts_pandascolumns(raw_dataset, 'Cohort ethnicity', cleaned=False)
value_counts_pandascolumns(raw_dataset, 'Allele', cleaned=False)
value_counts_pandascolumns(raw_dataset, 'Drug', cleaned=False)
```

```python
#ลบ ADR และยาที่บอกว่าจะตัดออก
ADR_toberemoved = ["DIHS","DILI","SOC","DISR","BIP","SOSG","AIA","AntiHGMCRmyopathy","AMNM","ALAT","Jaundice"]
Drug_toberemoved = ["EGF receptor inhibitors","tetanus antitoxin","antibacterial drugs","antiepileptic drugs","NSAID and 'multi-ingredient cold medication'","amoxicillin-clavulanate"]

#เลือกที่เกี่ยวข้อง
filtered_dataset = raw_dataset[['Drug','Allele','Cohort ethnicity','Pat1','ExpCtrl2','Adverse drug reaction']]
filtered_dataset = filtered_dataset.rename(columns={"Adverse drug reaction":"ADR"})
filtered_dataset = filtered_dataset.drop(filtered_dataset[filtered_dataset.ADR.isin(ADR_toberemoved)].index)
filtered_dataset = filtered_dataset.drop(filtered_dataset[filtered_dataset.Drug.isin(Drug_toberemoved)].index)
filtered_dataset
```
