```python
to_merge = pd.read_csv('1 cleaned_dataset_splited_output.csv')
```

```python
to_merge
```

```python
dataset_onlyadrs = to_merge.iloc[:,0:7].copy()
```

```python
adr_dict = {"No_ADR":0}
counter = 1
for x in dataset_onlyadrs.iloc[:,:9].columns:
    adr_dict[x] = counter
    counter += 1
adr_dict
```

```python
for lab,row in dataset_onlyadrs.iterrows():
    ADR = row['ADR']
    ADR_value = row[f"ADR_{ADR}"]
    if ADR_value == 1:
        dataset_onlyadrs.iloc[lab,adr_dict[ADR]] = adr_dict[f"ADR_{ADR}"]
    else:
        dataset_onlyadrs.iloc[lab,adr_dict[ADR]] = 0
```

```python
to_merge['ADR'] = dataset_onlyadrs['ADR']
to_merge
```

```python
to_merge = to_merge.drop(columns = list(adr_dict.keys()))
```
