```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('1 cleaned_dataset_splited_output.csv')
```

```python
#หา ADR ทั้งหมดที่มีใน input โดย loop แค่ 15 col แรกพอว่าอันไหนขึ้นต้นด้วย ADR_
ADR_inputed = []
for x in df.columns[0:16]:
    splited_x = x.split('_')
    print(f"{x}:{len(splited_x)}")
    try:
        if splited_x[0] == "ADR" and len(splited_x) == 2:
            ADR_inputed.append(x)
    except:
        continue
ADR_inputed
```

```python
#นำทุก columns ของ ADR มา sum กันใน axis 1 เพื่อบ่งบอกว่าเกิด หรือไม่เกิด ADR
df['ADR'] = df.loc[:,ADR_inputed].sum(axis=1)
```

```python
df
```

```python
df.drop(columns = ADR_inputed, inplace = True)
print(df.loc[:,['ADR','No_ADR']].sum(axis=0).sum()) #check ก่อน drop no_adr ว่าจำนวนที่เป็น 1 ยังเหมือนเดิมมั้ย (เท่ากับจำนวน rows มั้ย)
df.drop(columns = ['No_ADR'], inplace=True) #drop ทิ้งเพราะ no adr จะให้เป็น 0
```
